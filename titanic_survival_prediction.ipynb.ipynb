```python
# Titanic Problem

# Your job — build an ML model that predicts:

# "Given a passenger's details — did they survive or not?"



# Step 1 _ Load and explore data

import numpy as n
import pandas as p
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score,classification_report


df=p.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# print(df.shape)
# print(df.columns)
# print(df.info())
# print(df)

# Step 2 — Feature Engineering on real data:
df.drop(columns=['Name','Cabin','Ticket','PassengerId'],inplace=True)
# print(df)
df['Age'] = df['Age'].fillna(df['Age'].mean())

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
# print(df)
df['Sex']=df['Sex'].map({'male':0,'female':1})
df=p.get_dummies(df,columns=['Embarked'])
# print(df)

df['Family_Size']=df['SibSp']+df['Parch']+1

# df['isAlone']=df['Family'].apply(lambda x:1 if x==1 else 0)
df['Is_Alone'] = (df['Family_Size'] == 1).astype(int)

# print(df)


# print(df)
# print(df)

X=df.drop(columns=['Survived'])
Y=df['Survived']

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
pred=model.predict(x_test)

acc=accuracy_score(y_test,pred)
print(round(acc*100,2),"%")

cv_acc=cross_val_score(model,X,Y,cv=5)
print(round(cv_acc.mean()*100,2),"%")

important= {
    'features':X.columns,
    'imp':model.feature_importances_
}
# print(p.DataFrame(important))

new_pass = p.DataFrame({
    'Pclass':      [3],       # 3rd class
    'Sex':         [0],       # male
    'Age':         [25],      # 25 years old
    'SibSp':       [0],       # no siblings
    'Parch':       [0],       # no parents
    'Fare':        [7.25],    # cheap ticket
    'Embarked_C':  [0],
    'Embarked_Q':  [0],
    'Embarked_S':  [1],
    'Family_Size': [1],
    'Is_Alone':    [1]
})

new_pass1=new_passenger = p.DataFrame({
    'Pclass':      [1],       # change to 1st class
    'Sex':         [1],       # change to female
    'Age':         [25],
    'SibSp':       [0],
    'Parch':       [0],
    'Fare':        [100.0],   # expensive ticket
    'Embarked_C':  [1],
    'Embarked_Q':  [0],
    'Embarked_S':  [0],
    'Family_Size': [1],
    'Is_Alone':    [1]
})

# print(model.predict(new_pass1))
# if(model.predict(new_pass1)==1):
#   print("Survived")
# else:
#   print("Not Survived")

# Classification report
report=classification_report(y_test,pred,
                             target_names=['Survived','Not Survived'])
print(report)

```
